```python
# Import libraries

import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None



# Now we need to read in the data
df = pd.read_csv('/users/sunruolin/Desktop/study/movie_project/movies.csv')
```

```python
df.head()
```

```python
# find missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
df = df.dropna()
```

```python
# data type

df.dtypes
```

```python
# change data type
df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')

df.head()
```

```python
# create correct year column 
df['year_correct'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int)
```

```python
df
```

```python
df = df.sort_values(by=['gross'], inplace = False, ascending = False)
```

```python
# drop any duplicates

df.drop_duplicates()
```

```python
# budget will have high correlation 
# company will have correlation 

# scatter plot with budget VS gross

plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs Gross')
plt.xlabel('Budget')
plt.ylabel('Gross')
plt.show()
```

```python
# plot budget vs gross seaborn

sns.regplot(x='budget', y='gross', data = df, scatter_kws = {"color": "orange"}, line_kws = {"color": "blue"})
```

```python
# looking at correlation 

df.corr(method='pearson') #pearson(default), kendall
```

```python
# high correlation between budget and gross, it is right

correlation_matrix = df.corr(method='pearson')
sns.heatmap(correlation_matrix, annot = True)

plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movies Features')
plt.ylabel('Movies Features')
plt.show()
```

```python
df.head()
```

```python
df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized
```

```python
correlation_matrix_num = df_numerized.corr(method='pearson')
sns.heatmap(correlation_matrix_num, annot = True)

plt.title('Correlation Matrix for All Features')
plt.xlabel('Movies Features')
plt.ylabel('Movies Features')
plt.show()
```

```python
correlation_matrix_num
```

```python
corr_pairs = correlation_matrix_num.unstack()
corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5]
high_corr
```
